```python

import keras,os
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D , Flatten
from keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, initializers, Model, models, optimizers, metrics, callbacks, Input
import numpy as np
import ops
import tensorflow.train 
import matplotlib.pyplot as plt
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.activations import *
```

```python
train_dir="D:\\Data\\data\\train\\"
validation_dir="D:\\Data\\data\\test\\"
```

```python
BATCHSIZE=32
VALIDATION_SPLIT=0.1
WIDTH=600
HEIGHT=600
num_classes=2
train_ds=tf.keras.preprocessing.image_dataset_from_directory(
    "D:\\Data\\data_rgb\\train\\",
    labels="inferred",
    label_mode="int",
    color_mode="rgb",
    batch_size=BATCHSIZE,
    image_size=(HEIGHT, WIDTH),
    shuffle=True,
    seed=100,
    validation_split=VALIDATION_SPLIT,
    subset="training"
)
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "D:\\Data\\data_rgb\\test\\",
    subset="validation",
    seed=100,
    image_size=(HEIGHT, WIDTH),shuffle=True,
    validation_split=VALIDATION_SPLIT
)
    
```

```python
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
  for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.axis("off")
```

```python
trdata = ImageDataGenerator()
traindata = trdata.flow_from_directory(directory=train_dir,target_size=(224,224))
tsdata = ImageDataGenerator()
testdata = tsdata.flow_from_directory(directory=validation_dir, target_size=(224,224))
```

```python
from keras.applications.vgg16 import VGG16
model = VGG16()
```

```python
model.summary()
```

```python
num_classes = 2
img_height=600
img_width=600
model1 = Sequential([
  layers.experimental.preprocessing.Rescaling(1./255, input_shape=(img_height, img_width, 3)),
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(num_classes)
])
model1.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model1.summary()
```

```python
epochs=10
history = model1.fit(
  train_ds,
  validation_data=val_ds,
  epochs=epochs
)
```

```python
def create_image_list(path, labellist):
    with open('{}/image_list.txt'.format(path), 'w') as f:
        for label in labellist:
            for filename in os.listdir("{}\{}".format(path, label)):
                path_dir="{}\{}".format(path, label)
                filepath="{}\{}".format(path_dir, filename)
                f.write("{} {}\n".format(filepath, labellist.index(label)))
create_image_list("D:\\Data\\data\\train", ["non_ships", "ships"])

#https://stackoverflow.com/questions/34340489/tensorflow-read-images-with-labels
def read_labeled_image_list(image_list_file):
    f = open(image_list_file, 'r')
    filenames = []
    labels = []
    for line in f:
        filename, label = line[:-1].split(' ')
        filenames.append(filename)
        labels.append(int(label))
    return filenames, labels
def read_my_file_format(filename_and_label_tensor):
    filename =filename_and_label_tensor[0]
    label = filename_and_label_tensor[1]
    file_contents = tf.io.read_file(filename)
    example = tf.io.decode_raw(file_contents, tf.uint16)
    return example, label
```
